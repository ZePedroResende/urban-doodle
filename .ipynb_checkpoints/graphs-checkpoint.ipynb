```python
from scipy.cluster.hierarchy import linkage, dendrogram
import pandas as pd
import numpy as np
import networkx as nx
import  itertools as it
import matplotlib.pyplot as plt
from scipy.cluster import hierarchy as hc
import re
```

```python
#data = arff.loadarff("datasets/treated.csv")

#df = pd.DataFrame(data[0])
from scipy.spatial.distance import pdist
df = pd.read_csv("datasets/treated.csv", na_values=["?"])
df.head()
```

```python
df.replace('?', np.NaN)
#df['rbc'].replace(df['rbc'].head()[0], inplace=True)


df.isnull().values.any()
df.isnull().sum()
col = df.columns.tolist()
```

```python
corr = 1 - df.corr() 

corr_condensed = hc.distance.squareform(corr)
z = hc.linkage(corr_condensed, method='average')
hc.dendrogram(z, labels=corr.columns)
plt.show()
```

```python
graph = nx.Graph()


for index, row in df.iterrows():
    for i in it.combinations(col,2):
        if not pd.isnull(row[i[0]]) and not pd.isnull( row[i[1]]):
            node_a = i[0]+ str(row[i[0]])
            node_b = i[1]+ str(row[i[1]])
        
            if graph.has_edge(node_a, node_b):
                    graph[node_a][node_b]['weight'] += 1
            else:
                    graph.add_edge(node_a,node_b, weight=1)
        

nx.draw(graph, layout=nx.spring_layout(graph,  k=0.3*1/np.sqrt(len(graph.nodes())), iterations=20), scale=25)
plt.figure(3, figsize=(300, 300))

plt.show()
```

```python
nx.draw_networkx(graph, pos=nx.random_layout(graph), width=[d['weight'] for _, _, d in graph.edges(data=True)])
plt.show()
```

```python
G = graph
edges,weights = zip(*nx.get_edge_attributes(G,'weight').items())

pos = nx.spring_layout(G)
nx.draw(G, pos, node_color='b', edgelist=edges, edge_color=weights, width=10.0)
plt.figure(3,figsize=(100, 100))
plt.show()
```

```python
node_list = graph.number_of_nodes()
all_weights = []
for (node1,node2,data) in graph.edges(data=True):
    all_weights.append(data['weight']) #we'll use this when determining edge thickness
unique_weights = list(set(all_weights))

for weight in unique_weights:
    weighted_edges = [(node1,node2) for (node1,node2,edge_attr) in G.edges(data=True) if edge_attr['weight']==weight]
    width = weight
    nx.draw(G,pos,edgelist=weighted_edges,width=width)

width = (weight*node_list)/sum(all_weights)

plt.show()
```
